# T-Test for Difference in Means

Project description

In this project, we aim to explore the differences in spending behavior between male and female customers using a dataset obtained from a mall customer analysis. The primary objective is to determine whether there is a statistically significant difference in the spending scores assigned to these two groups using an independent t-test.

Data Preparation: The dataset is filtered to create two separate groups based on gender:

Male Group: Contains spending scores of male customers.
Female Group: Contains spending scores of female customers.

Statistical Analysis: An independent t-test is performed to assess whether there is a significant difference in the spending scores between the two groups. The t-test produces:
- T-statistic: A value indicating the ratio of the difference between the group means to the variability of the groups.
- P-value: A probability value that helps determine the statistical significance of the results.

Significance Testing: A significance level (alpha) of 0.05 is set. If the p-value obtained from the t-test is less than this threshold, it indicates that the difference between the spending scores of males and females is statistically significant.

```python
import pandas as pd
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Mall_Customers.csv')

# Split data into groups by gender
male_group = data[data['Genre'] == 'Male']['Spending Score (1-100)']
female_group = data[data['Genre'] == 'Female']['Spending Score (1-100)']

# Perform the independent t-test
t_stat, p_val = ttest_ind(male_group, female_group)
# Print the t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_val}")

# Check for statistical significance
alpha = 0.05  # Significance level
if p_val < alpha:
    print("The difference between groups is significant")
else:
    print("The difference between groups is not significant")

# Visualization: Create a boxplot to show spending scores by gender
sns.boxplot(x='Genre', y='Spending Score (1-100)', data=data)
plt.title("Spending Score by Gender")
plt.show()
```
